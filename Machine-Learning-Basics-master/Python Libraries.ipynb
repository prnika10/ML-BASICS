### Matplotlib

Plotting library for Python

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,4,6,8])
y = x**2 + 5
```

```python
print(x)
print(y)
```

```python
plt.plot(x,y)
```

```python
plt.style.use('classic')
```

```python
plt.plot(x,y,label = "House Prices")
plt.title("House Pricing Graph")
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x,y,label = "House Prices")
plt.plot(x+2,y+3,label = "House Prices-1")
plt.title("House Pricing Graph")
plt.legend()
plt.show()
```

```python
plt.scatter(x, x**2 + 1, label = "House Price 1")
plt.scatter(x, x + 10, label = "House Price 2")
```

#### OpenCV

Library for real-time computer vision

```python
import cv2
```

```python
pic = cv2.imread('flower.jpg')
plt.imshow(pic)
```

```python
print(pic.shape)
```

Shows the number of pixels and channels. There are three channels - RGB.

But OpenCV reads image in BGR format and hence we'll need to convert it.

```python
new_pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
plt.imshow(new_pic)
plt.axis('off')
plt.show()
```

We can slice the image to crop it.

```python
crop_pic = new_pic[40:620,180:750,:]
plt.imshow(crop_pic)
```

```python
crop_pic.shape
```

```python
crop_pic = cv2.resize(crop_pic, (102, 100))
plt.imshow(crop_pic)
plt.axis('off')
plt.show()
```

```python
crop_pic.shape
```

The new image has lower resolution and has been resized.
